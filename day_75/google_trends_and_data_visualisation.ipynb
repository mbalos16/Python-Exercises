### Notebook Formatting & Style Helpers

```python
# Create locators for ticks on the time axis
years = mdates.YearLocator()
months = mdates.MonthLocator()
years_fmt = mdates.DateFormatter( "%Y")
```

### Tesla Stock Price v.s. Search Volume

```python
x = df_tesla["MONTH"]
y = df_tesla["TSLA_USD_CLOSE"]
y2 = df_tesla["TSLA_WEB_SEARCH"]
title = "Tesla Web Search vs Price"

plt.figure(figsize = (14, 8))
ax1 = plt.gca()  # get the axis
ax2 = ax1.twinx()  # create another axis that shares the same axis
ax1.plot(x, y, color = "b")  # plot the first graph
ax2.plot(x, y2, color = "g")  # plot the second graph
ax1.set_xlabel("Year", fontsize = 12, color = "purple")  # set the x label
ax1.set_ylabel("TSLA Stock Price", fontsize = 12, color = "blue")  # set the left y label
ax2.set_ylabel("Search Trend", fontsize = 12, color = "green")  # set the right y label
plt.title(title, fontsize = 16)  # add title
ax1.set_ylim([0, 600])  # set the y ax limit
ax1.set_xlim([x.min(), x.max()])  # set the x ax limit
ax1.xaxis.set_major_locator(years) # add ticks in the graph
ax1.xaxis.set_major_formatter(years_fmt) # add ticks in the graph
ax1.xaxis.set_minor_locator(months)  # add ticks in the graph
ax1.grid(True)  # add grid to the graph
plt.show();
```

### Unemployement Benefits Search vs. Actual Unemployment in the U.S.
