#### 1-Dimensional Arrays (Vectors)

```python
# Create an array
my_array = np.array([1.1, 9.2, 8.1, 4.7])
```

```python
# Show the shape of the array
my_array.shape
```

```python
# Access the second element of an array
my_array[2]
```

```python
# Check the dimmensions of the array
my_array.ndim
```

#### 2-Dimensional Arrays (Matrices)

```python
# Creation of a 2D array
array_2d = np.array([[1, 2, 3, 9], 
                     [5, 6, 7, 8]])
```

```python
print(f"The array_2d has {array_2d.ndim} dimensions.")
print(f"It's shape is: {array_2d.shape}.")
print(f"It has {array_2d[0]} rows and {array_2d[1]} columns.")
print(array_2d)
```

```python
#Access the third value in the second row of the 2D array:
array_2d[1][2]
```

```python
# or 
array_2d[1, 2]
```

```python
# Access the entire 1st row of the 2d array:
array_2d[0, :]
```

```python
# Access the entire 2nd row of the 2d array:
array_2d[1, :]
```

```python
img = misc.face()
```

**Challenge**: What is the data type of `img`? Also, what is the shape of `img` and how many dimensions does it have? What is the resolution of the image?

```python
# Check for image resolution
hgt = img.shape[0]
wid = img.shape[1]

print(f"The resolution of the image is width = {wid} and height = {hgt} and it's resolution is {wid*hgt}.")
```

**Challenge**: Convert the image to black and white. The values in our `img` range from 0 to 255. 
* Divide all the values by 255 to convert them to sRGB, where all the values are between 0 and 1. 
* Next, multiply the sRGB array by the `grey_vals` to convert the image to grey scale. 
* Finally use Matplotlib's [`.imshow()`](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.imshow.html) together with the colormap parameter set to gray `cmap=gray` to look at the results. 

```python
grey_vals = np.array([0.2126, 0.7152, 0.0722])
```

```python
sRGB = img / 255
```

```python
grayscale = np.matmul(sRGB, grey_vals)
```

```python
plt.imshow(grayscale, cmap = "gray")
plt.show()
```

**Challenge**: Can you manipulate the images by doing some operations on the underlying ndarrays? See if you can change the values in the ndarray so that:

#### Challenge Solutions

```python
# Flip the image
flipped_img = np.flip(grayscale)
plt.imshow(flipped_img, cmap = "gray")
plt.show()
```

```python
# Rotate the color image
plt.imshow(np.rot90(img))
plt.show()
```

```python
solar_img = 255 - img
plt.imshow(solar_img)
plt.show()
```
